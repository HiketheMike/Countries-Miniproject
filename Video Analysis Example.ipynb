## Long-Run Analysis

The long run deals with structural economic definitions, structural neutral rates, anchored expectations, and the ultimate goals of the dual mandate.

* **The Ultimate Target:** The price stability leg of the Federal Reserve's dual mandate is fundamentally defined as achieving a 2% annual headline PCE growth over the long term. Long-term trends are considered the vital benchmark for policy because monetary policy operates with long and variable lags.
* **Well-Anchored Expectations:** Despite recent periods of excess inflation and tariff uncertainty, long-term market and survey-based inflation expectations have remained firmly anchored. This anchors the public's trust in forward guidance and enhances the Fed's long-term capability to fulfill its dual mandate.
* **Long-Run Balance Sheet Goals:** Over the past three years, the Fed has utilized quantitative tightening to transition its balance sheet from an abundant reserves regime to an ample reserves regime without distorting the long-term fixed-income market. 
* **Structural Long-Run R:** Long-run R* represents the structural intersection between the supply of savings and the demand for investment, completely independent of short-term economic headwinds or tailwinds. It dictates the structural "breathing room" the Fed has; a higher long-run R* grants policymakers more ammunition to implement expansionary policy before being constrained by the zero/effective lower bound.
